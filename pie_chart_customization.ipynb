```python
from matplotlib import pyplot as plt
```

```python
slices = [240,60]
labels = ['first','secound']
plt.pie(slices,labels = labels)
plt.tight_layout()
plt.title('first pie chart')
plt.show()
```

```python
plt.style.use("fivethirtyeight")

slices1 = [240,60,40,30]
labels = ['first','secound','fouth','fifth']
color = ['blue','green','yellow','red']
#['#CD5C5C','#DFFF00','#454545','#0000FF']
plt.pie(slices1,labels = labels,colors=color , wedgeprops={'edgecolor':'green'})

plt.tight_layout()
plt.title('first pie chart')
plt.show()
```

```python
# Language Popularity
useno = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
plt.pie(useno,labels = language)

plt.tight_layout()
plt.title('Language Popularity')
plt.show()
```

```python
#for explode 
useno = [59219, 55466, 47544, 36443, 35917]
language = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0.2, 0.1,  0]
plt.pie(useno,labels = language,explode=explode,shadow=True,startangle=90,autopct='%1.1f%%')

plt.tight_layout()
plt.title('Language Popularity')
plt.show()
```
