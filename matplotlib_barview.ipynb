```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['google','amazon','olx','facebook']
revenue=[23,100,50,60]
profit=[20,30,10,20]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue")
plt.title('Stocks Price')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='revenue')
plt.ylabel('revenue')
plt.title('stock price')
plt.legend()
```

```python
plt.barh(xpos,revenue, label="Revenue") 

plt.yticks(xpos,company)
plt.ylabel("Revenue")
plt.title('Stocks Price')
plt.legend()
```
