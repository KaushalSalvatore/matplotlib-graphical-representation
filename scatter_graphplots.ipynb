```python
import pandas as pd
from matplotlib import pyplot as plt

```

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x, y, s=100, c='red', linewidth=1, alpha=0.75 , edgecolor='green' )#s for dot size , c for color change , marker 


plt.tight_layout()

plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5] #default color dots
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]#custom dot size
plt.scatter(x, y, s=sizes, c=colors , cmap='Greens' ,linewidth=1, alpha=0.75 , edgecolor='green' ) # cmap for dot color customzing

cbar = plt.colorbar()
cbar.set_label('measurement')

plt.tight_layout()

plt.show()
```

```python

data = pd.read_csv('scatterplots.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes ,linewidth=1, c=ratio, cmap='summer', alpha=0.75 , edgecolor='green' )
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
