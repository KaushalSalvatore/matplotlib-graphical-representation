```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
max_stu=[23,24,25,56,67,78,30]
min_stu=[13,14,15,36,47,58,20]
avg_stu=[10,20,22,30,35,40,15]
```

```python
plt.plot(x,max_stu,label='max')
plt.plot(x,avg_stu,label='avg')
plt.plot(x,min_stu,label='min')
plt.legend(shadow=True,fontsize='small')# set legend legend(loc='lower left')//(loc='best') 
plt.xlabel('days')
plt.ylabel('student')
plt.title('Attendance')
plt.grid()
```
