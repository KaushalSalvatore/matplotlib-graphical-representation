```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7]
y=[23,35,25,56,67,78,30]
```

```python
plt.plot(x,y,'rD')
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='dotted',markersize='5') #we can also use color code 
```

```python
plt.plot(x,y,color='green',marker='+',alpha=0.5)# transparency lines alpha value 0 to 1 
```
