```python
# simple chart view between student or class attendance
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7]
y=[23,24,25,56,67,78,30]
```

```python
plt.xlabel('Day')
plt.ylabel('Student')
plt.title('attendance')
plt.plot(x,y,color='green',linewidth=2,linestyle='dashdot')
```
